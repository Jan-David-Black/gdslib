# Circuit simulations

Simphony can also chain components into circuits

```python
import matplotlib.pyplot as plt
import pp
import gdslib as gl

circuit = gl.mzi(L0=1, DL=100, L2=10, y_model_factory=gl.mmi1x2)

gl.sweep_simulation(
    circuit,
    iport="input",
    oport="output",
    start=1500e-9,
    stop=1600e-9,
    logscale=True,
)
```

```python
circuit_gc = gl.add_gc(circuit)

gl.sweep_simulation(
    circuit_gc,
    iport="input",
    oport="output",
    start=1500e-9,
    stop=1600e-9,
    logscale=True,
)
```

# Montecarlo variation

We can also account for montecarlo variations of the circuits

```python
gl.sweep_simulation_montecarlo(
    circuit_gc,
    iport="input",
    oport="output",
    start=1500e-9,
    stop=1600e-9,
    logscale=True,
    runs=10
)
```
